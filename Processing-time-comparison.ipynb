```python
import torch
from torchvision import models
from torchvision.models import EfficientNet_B7_Weights

# PyTorch
pytorch_model = models.efficientnet_b7(weights=EfficientNet_B7_Weights.IMAGENET1K_V1)
pytorch_model.eval()
torch.save(pytorch_model, 'efficientnet_b7.pth')
```

```python
# PyTorch -> ONNX
torch.onnx.export(
    pytorch_model,
    torch.randn(1, 3, 224, 224),
    'efficientnet_b7.onnx',
    input_names=['input'],
    output_names=['output'],
)
```

```python
import openvino as ov

# ONNX -> OpenVINO
ov_model = ov.convert_model('efficientnet_b7.onnx')
ov.save_model(ov_model, 'efficientnet_b7.xml')
```

# サンプル画像取得

```python
!pip install -q imread_from_url
```

```python
from imread_from_url import imread_from_url

image = imread_from_url('https://raw.githubusercontent.com/Kazuhito00/OpenVINO-Colab-Inference-Sample-using-various-frameworks-weights/main/sample.jpg')
```

```python
import cv2
import numpy as np

def preprocess_image(image):
    preprocessed_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    preprocessed_image = cv2.resize(preprocessed_image, (224, 224))

    preprocessed_image = preprocessed_image.astype(np.float32) / 255.0
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    preprocessed_image = (preprocessed_image - mean) / std

    bhwc_image = np.expand_dims(preprocessed_image, axis=0).astype(np.float32)

    bchw_image = np.transpose(preprocessed_image, (2, 0, 1))
    bchw_image = np.expand_dims(bchw_image, axis=0).astype(np.float32)

    return bchw_image, bhwc_image

# 前処理
bchw_image, bhwc_image = preprocess_image(image)
torch_image = torch.tensor(bchw_image, dtype=torch.float32)
```

# 重み読み込み

```python
# PyTorch
import torch

pytorch_model = torch.load('efficientnet_b7.pth', map_location=torch.device('cpu'))
pytorch_model.eval()
```

```python
%%timeit -n 10 -r 5

with torch.no_grad():
    _ = pytorch_model(torch_image)
```

```python
# ONNX
import onnxruntime

onnx_model = onnxruntime.InferenceSession('efficientnet_b7.onnx', providers=['CPUExecutionProvider'])
```

```python
%%timeit -n 10 -r 5

_ = onnx_model.run(None, {"input": bchw_image})
```

```python
# TensorFlow
import tensorflow as tf

with tf.io.gfile.GFile('efficientnet_b7.pb', "rb") as f:
    graph_def = tf.compat.v1.GraphDef()
    graph_def.ParseFromString(f.read())

with tf.Graph().as_default() as graph:
    tf.import_graph_def(graph_def, name="")

tf_input_name = graph.get_tensor_by_name('input:0')
tf_output_name = graph.get_tensor_by_name('Identity:0')

tf_model = tf.compat.v1.Session(graph=graph)
```

```python
%%timeit -n 10 -r 5

_ = tf_model.run(tf_output_name, feed_dict={tf_input_name: bhwc_image})
```

```python
# TensorFlow Lite

tflite_model = tf.lite.Interpreter(model_path='efficientnet_b7.tflite')
tflite_model.allocate_tensors()

input_details = tflite_model.get_input_details()
```

```python
%%timeit -n 10 -r 5

tflite_model.set_tensor(input_details[0]['index'], bhwc_image)
tflite_model.invoke()
```

```python
# PaddlePaddle
import paddle

paddle_model = paddle.jit.load('efficientnet_b7')
paddle_model.eval()
```

```python
%%timeit -n 10 -r 5

with paddle.no_grad():
    _ = paddle_model(bchw_image)
```

```python
# OpenVINO
import openvino as ov

core = ov.Core()

ov_openvino_model = core.read_model(model='efficientnet_b7.xml')
ov_openvino_model = core.compile_model(model=ov_openvino_model, device_name="CPU")

ov_onnx_model = core.read_model(model='efficientnet_b7.onnx')
ov_onnx_model = core.compile_model(model=ov_onnx_model, device_name="CPU")

ov_tf_model = core.read_model(model='efficientnet_b7.pb')
ov_tf_model = core.compile_model(model=ov_tf_model, device_name="CPU")

ov_tflite_model = core.read_model(model='efficientnet_b7.tflite')
ov_tflite_model = core.compile_model(model=ov_tflite_model, device_name="CPU")

ov_paddle_model = core.read_model(model='efficientnet_b7.pdmodel')
ov_paddle_model = core.compile_model(model=ov_paddle_model, device_name="CPU")
```

```python
%%timeit -n 10 -r 5

_ = ov_openvino_model([bchw_image])
```

```python
%%timeit -n 10 -r 5

_ = ov_onnx_model([bchw_image])
```

```python
%%timeit -n 10 -r 5

_ = ov_tf_model([bhwc_image])
```

```python
%%timeit -n 10 -r 5

_ = ov_tflite_model([bhwc_image])
```

```python
%%timeit -n 10 -r 5

_ = ov_paddle_model([bchw_image])
```
